```python
import os
import random
import shutil
import zipfile

# ----------------------------
# CONFIG
# ----------------------------
folders = [
    "/kaggle/input/fruit-and-vegetable-disease-healthy-vs-rotten/Fruit And Vegetable Diseases Dataset/Bellpepper__Healthy",
    "/kaggle/input/fruit-and-vegetable-disease-healthy-vs-rotten/Fruit And Vegetable Diseases Dataset/Bellpepper__Rotten",
    "/kaggle/input/fruit-and-vegetable-disease-healthy-vs-rotten/Fruit And Vegetable Diseases Dataset/Carrot__Healthy",
    "/kaggle/input/fruit-and-vegetable-disease-healthy-vs-rotten/Fruit And Vegetable Diseases Dataset/Carrot__Rotten",
    "/kaggle/input/fruit-and-vegetable-disease-healthy-vs-rotten/Fruit And Vegetable Diseases Dataset/Cucumber__Healthy",
    "/kaggle/input/fruit-and-vegetable-disease-healthy-vs-rotten/Fruit And Vegetable Diseases Dataset/Cucumber__Rotten",
    "/kaggle/input/fruit-and-vegetable-disease-healthy-vs-rotten/Fruit And Vegetable Diseases Dataset/Mango__Healthy",
    "/kaggle/input/fruit-and-vegetable-disease-healthy-vs-rotten/Fruit And Vegetable Diseases Dataset/Tomato__Healthy",
    "/kaggle/input/fruit-and-vegetable-disease-healthy-vs-rotten/Fruit And Vegetable Diseases Dataset/Tomato__Rotten"
]

IMAGES_PER_FOLDER = 200
OUTPUT_BASE = "/kaggle/working/selected_images"
ZIP_PATH = "/kaggle/working/selected_images.zip"

# ----------------------------
# CREATE OUTPUT FOLDER
# ----------------------------
os.makedirs(OUTPUT_BASE, exist_ok=True)

# ----------------------------
# RANDOMLY PICK & COPY IMAGES
# ----------------------------
for folder in folders:
    class_name = os.path.basename(folder)
    dest_folder = os.path.join(OUTPUT_BASE, class_name)
    os.makedirs(dest_folder, exist_ok=True)

    images = [
        f for f in os.listdir(folder)
        if f.lower().endswith((".jpg", ".jpeg", ".png"))
    ]

    if len(images) < IMAGES_PER_FOLDER:
        raise ValueError(f"Not enough images in {folder}")

    selected_images = random.sample(images, IMAGES_PER_FOLDER)

    for img in selected_images:
        shutil.copy(
            os.path.join(folder, img),
            os.path.join(dest_folder, img)
        )

    print(f"✔ Copied {IMAGES_PER_FOLDER} images from {class_name}")

# ----------------------------
# ZIP EVERYTHING
# ----------------------------
with zipfile.ZipFile(ZIP_PATH, "w", zipfile.ZIP_DEFLATED) as zipf:
    for root, _, files in os.walk(OUTPUT_BASE):
        for file in files:
            full_path = os.path.join(root, file)
            arcname = os.path.relpath(full_path, OUTPUT_BASE)
            zipf.write(full_path, arcname)

print("\n✅ ZIP file created:", ZIP_PATH)
```
